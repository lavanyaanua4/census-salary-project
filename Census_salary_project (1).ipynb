Q1. which age group people earn more per hour?

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()
sns.scatterplot(x='age', y='wage_per_hour',data=df)
plt.title("wage per hour efffect by age")
plt.show()
```

Q2. Which marital status group earns the most?

```python
plt.figure()
sns.barplot(x='marital_status',y='wage_per_hour',data=df)
plt.title("wage per hour as per marital status")
plt.xticks(rotation=90)
plt.show()
```

Q3. Which race group has the highest average wage?


Q9. Top 10 household roles by average wage

```python
df
```

```python
a=df.groupby('household_stat')['wage_per_hour'].mean().sort_values(ascending=False).head(10).reset_index()
plt.figure()
sns.barplot(x='household_stat',y='wage_per_hour',data=a)
plt.title("wage per hour based on household_stat")
plt.xticks(rotation=90)
plt.show()
```

Q10. Are certain education levels more common among high-income people?
